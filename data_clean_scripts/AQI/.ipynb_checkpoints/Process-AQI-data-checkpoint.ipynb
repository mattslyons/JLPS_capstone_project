# Processing the EPA AQI Data

Process the full AQI, PM2.5 data from the EPA website.

Take daily measurements from 1999 - 2017, 

aggregate into monthly measurements, and save it off.

aggregate into weekly measurements, and save it off.


Reference for the dataset: https://www.epa.gov/outdoor-air-quality-data/about-air-data-reports


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
import os
import re 


pd.set_option('display.max_columns', None)
```

## Load in the EPA daily data

```python
path_header = '../../data/EPA_AQI_Data/'
fname = 'ca_aqi_epa_1999_daily.csv'

fpath = path_header + fname
```

```python
df = pd.read_csv(fpath)
```

```python
df
```

```python
df.columns
```

```python
df.isna().sum()
```

## Load in the other daily datasets and combine in

```python
# os.chdir(fpath)
# data_files = [i for i in os.listdir(path) if re.search('Data.csv', i)]
data_files = [f'../../data/EPA_AQI_Data/ca_aqi_epa_{i}_daily.csv' for i in range(2000, 2018)]
data_files
```

```python
for data_fname in data_files:
    df_loaded = pd.read_csv(data_fname)
    df = pd.concat([df, df_loaded], axis=0, ignore_index=False)
del df_loaded
```

```python
display(df)
df.columns
```

```python
df.isna().sum()
```

## Clean up the big dataframe

```python
# fill in the missing Site Names with 'unknown'

df['Site Name'] = df['Site Name'].fillna('unknown')

df.isna().sum()
```

```python
# convert the 'Date' column to a datetime object

df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

## Aggregate daily observations b/c of duplicates

We care about aggregating daily observations for a site based on the `'Date', 'Site ID', 'SITE_LATITUDE', 'SITE_LONGITUDE'`

```python
df.describe()
```

```python
df.columns
```

```python
cols_to_groupby = ['Date', 'Source', 'Site ID',
       'UNITS', 'Site Name',
       'PERCENT_COMPLETE', 
       'STATE_CODE', 'STATE', 'COUNTY_CODE',
       'COUNTY', 'SITE_LATITUDE', 'SITE_LONGITUDE']

cols_to_aggregate = ['Daily Mean PM2.5 Concentration', 'DAILY_AQI_VALUE', 'DAILY_OBS_COUNT']
```

```python
# aggregate the data daily

# AQI should be averaged across the day
# PM2.5 should be averaged across the day
# daily observation count should be summed in the day

df_daily = df.groupby(cols_to_groupby, as_index=False).agg(
    AQI_daily_mean=pd.NamedAgg(column='DAILY_AQI_VALUE', aggfunc=np.mean), 
    PM2point5_daily_mean=pd.NamedAgg(column='Daily Mean PM2.5 Concentration', aggfunc=np.mean),
    daily_obs_count=pd.NamedAgg(column='DAILY_OBS_COUNT', aggfunc=np.sum)
)

df_daily
```

# Add in new columns for year, month, week number

```python
df_daily['year'] = pd.DatetimeIndex(df_daily['Date']).year
df_daily['month'] = pd.DatetimeIndex(df_daily['Date']).month

# year-month
df_daily['year_month'] = df_daily['Date'].dt.to_period('M')

# week number
df_daily['week_number'] = df_daily['Date'].dt.week
# year-week number
df_daily['year_week'] = df_daily['Date'].dt.strftime('%Y-%W')
```

```python
display(df_daily)
```

### Save daily dataset to CSV

```python
fpath = path_header + 'ca_aqi_epa_1999-2017_processed_daily.csv.csv'
```

```python
df_daily.to_csv(path_or_buf=fpath, index=False)
```

## Monthly Aggregation

```python
df_daily.columns
```

```python
df_daily.head()
```

```python
df_monthly = df_daily.groupby(['year','month', 'year_month', 'Site ID', 'STATE_CODE', 'STATE', 'COUNTY_CODE',
       'COUNTY', 'Site Name', 'SITE_LATITUDE', 'SITE_LONGITUDE', 'UNITS'], as_index=False).agg(
           AQI_monthly_mean=pd.NamedAgg(column='AQI_daily_mean', aggfunc=np.mean), 
           PM2point5_monthly_mean=pd.NamedAgg(column='PM2point5_daily_mean', aggfunc=np.mean),
           obs_count_monthly =pd.NamedAgg(column='daily_obs_count', aggfunc=np.sum)
           )
       
```

```python
df_monthly['month_year_key']  = df_monthly["year_month"].astype(str) + '-ym-' + df_monthly["Site ID"].astype(str)
```

```python
display(df_monthly)
```

```python
len(pd.unique(df_monthly.month_year_key))
```

```python
# EDA on monthly data
df_monthly.describe()
```

```python
# AQI
display(df_monthly.AQI_monthly_mean.hist(bins=50))
```

```python
# PM2.5 daily mean
display(df_monthly.PM2point5_monthly_mean.hist(bins=50))
```

```python
# PM2.5 daily mean
display(df_monthly.obs_count_monthly.hist(bins=50))
```

### Save monthly dataset to CSV

```python
fpath = path_header + 'ca_aqi_epa_1999-2017_monthly.csv'
```

```python
df_monthly.to_csv(path_or_buf=fpath, index=False)
```

## Weekly Aggregation
