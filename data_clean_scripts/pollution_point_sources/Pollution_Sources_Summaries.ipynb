# NEI Zip Code Truncation (complete; don't run again)

```python
def truncate_zip(zipcode_str):
    try:
        zipcode_str = str(zipcode_str)
        if len(zipcode_str) > 5:
            if zipcode_str[0] == "'" and zipcode_str[-1] == "'":
                zipcode_str = zipcode_str.strip("'")
            else:
                zipcode_str = int(zipcode_str[:5])
    except ValueError:
        zipcode_str = np.NAN
    return zipcode_str
```

```python
nei1999['zip_code'] = nei1999['Location Zip Code'].apply(truncate_zip)
```
