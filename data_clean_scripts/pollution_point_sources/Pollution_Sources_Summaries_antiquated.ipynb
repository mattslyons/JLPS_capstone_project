```python
def get_records(df):
    df = df
    return df.shape[0]

def get_columns(df):
    df = df
    cols = list(df.columns)
    return cols

def get_unique(df, col_name):
    df = df
    col_name = col_name
    return len(df[col_name].unique())

def count_unique_column_values(df, col_name):
    df = df
    col_name = col_name
    print("Number unique records for {} is {}".format(col_name, get_unique(df, col_name)))
          
for i in df_list:
    records = get_records(i)
    columns = get_columns(i)
    print('''
    ###########################################################################
                                       {}
    ###########################################################################
    '''.format(i.name))
    print("Number of records for {} is {}\n\n".format(i.name, records))
    print("Columns for {} are\n {}\n\n".format(i.name, columns))
    if i.name in ['nei1999', 'nei2002', 'nei2005']:
#         for j in i.columns:
#             count_unique_column_values(i, j)
        print("Number of unique entries for {} is {}".format('NTI Unique Facility ID', get_unique(i, 'NTI Unique Facility ID')))
        print("Number of unique entries for {} is {}".format('Facility Site Name', get_unique(i, 'Facility Site Name')))
    elif i.name in ['nei2008', 'nei2011', 'nei2014']:
#         for j in i.columns:
#             count_unique_column_values(i, j)
        print("Number of unique entries for {} is {}".format('eis_facility_site_id', get_unique(i, 'eis_facility_site_id')))
        print("Number of unique entries for {} is {}".format('facility_site_name', get_unique(i, 'facility_site_name')))
        print("Number of unique entries for {} is {}".format('location_address_text', get_unique(i, 'location_address_text')))
        print("Number of unique entries for {} is {}".format('state_and_county_fips_code', get_unique(i, 'state_and_county_fips_code')))
        print("Number of unique entries for {} is {}".format('scc', get_unique(i, 'scc')))
    
    elif i.name in ['nei2017']:
        #i.hist(column='total_emissions', bins=20, log=True)
#         for j in i.columns:
#             count_unique_column_values(i, j) 
        print("Number of unique entries for {} is {}".format('eis_facility_id', get_unique(i, 'eis_facility_id')))
        print("Number of unique entries for {} is {}".format('site_name', get_unique(i, 'site_name')))
        print("Number of unique entries for {} is {}".format('naics_code', get_unique(i, 'naics_code')))
        print("Number of unique entries for {} is {}".format('agency_facility_id', get_unique(i, 'agency_facility_id')))
        print("Number of unique entries for {} is {}".format('address', get_unique(i, 'address')))
        print("Number of unique entries for {} is {}".format('fips', get_unique(i, 'fips')))
        print("Number of unique entries for {} is {}".format('scc', get_unique(i, 'scc')))
        print("Number of unique entries for {} is {}".format('eis_unit_id', get_unique(i, 'eis_unit_id')))
        
        
```

# NEI Zip Code Truncation (complete; don't run again)

```python
def truncate_zip(zipcode_str):
    try:
        zipcode_str = str(zipcode_str)
        if len(zipcode_str) > 5:
            if zipcode_str[0] == "'" and zipcode_str[-1] == "'":
                zipcode_str = zipcode_str.strip("'")
            else:
                zipcode_str = int(zipcode_str[:5])
    except ValueError:
        zipcode_str = np.NAN
    return zipcode_str
```

```python
nei1999['zip_code'] = nei1999['Location Zip Code'].apply(truncate_zip)
```
